```python
import pandas as pd
import numpy as np

# Load csv
df = pd.read_csv('clean_data/data_cleaned.csv')
df.describe()
```

```python
df
```

```python
targets = df['close'].shift(-1)
targets
```

```python
# Remove last row to match targets
df = df[:-1]
targets = targets[:-1]
```

```python
# Drop date column
df = df.drop(columns=['date'])
```

```python
# Split dataset and targets into train, validation and test with 50-25-25 ratio
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, targets, test_size=0.5, shuffle=False)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, shuffle=False)

X_train.shape, X_val.shape, X_test.shape
```

```python
# Convert datasets to numpy arrays
X_train = X_train.to_numpy()
X_val = X_val.to_numpy()
X_test = X_test.to_numpy()

y_train = y_train.to_numpy()
y_val = y_val.to_numpy()
y_test = y_test.to_numpy()
```

# Classification

```python
# Convert targets to classification targets
y_train_classification = np.where(y_train > 0, 1, 0)
y_val_classification = np.where(y_val > 0, 1, 0)
y_test_classification = np.where(y_test > 0, 1, 0)
```

```python
# Apply standard scaler to the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_val_scaled = scaler.transform(X_val)
```

```python
# Logistic regression model
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(X_train_scaled, y_train_classification)

# Predictions
y_pred = model_lr.predict(X_val_scaled)

# Accuracy
from sklearn.metrics import accuracy_score

accuracy_score(y_val_classification, y_pred)
```

```python
# Try support vector machines with linear, radial, polynomial kernel
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt

kernels = ['linear', 'rbf', 'poly']
gamma = [0.1, 0.5, 1, 5, 10, 100]
C = [0.1, 1, 10, 100]

# For each C, make a heatmap of the accuracy of each model trained with a different kernel and gamma
for c in C:
    accuracies = np.zeros((len(kernels), len(gamma)))
    for i in range(len(kernels)):
        for j in range(len(gamma)):
            model = SVC(kernel=kernels[i], gamma=gamma[j], C=c)
            model.fit(X_train_scaled, y_train_classification)
            y_pred = model.predict(X_val_scaled)
            accuracies[i, j] = accuracy_score(y_val_classification, y_pred)
    plt.figure(figsize=(8, 4))
    sns.heatmap(accuracies, annot=True, xticklabels=gamma, yticklabels=kernels)
    plt.title(f'Accuracy for C={c} and kernels')
    plt.xlabel('Gamma')
    plt.ylabel('Kernel')
    plt.show()
```

```python
# Build many random forest classifiers with the following parameters
from sklearn.ensemble import RandomForestClassifier

n_trees = [500, 1000, 2000]
max_depth = [5, 10, 20, 50, 100]
sampling = [0.5, 0.3]

# For each number of trees, make a heatmap of the accuracy of each model trained with a different max depth and sampling
for n in n_trees:
    accuracies = np.zeros((len(max_depth), len(sampling)))
    for i in range(len(max_depth)):
        for j in range(len(sampling)):
            model = RandomForestClassifier(n_estimators=n, max_depth=max_depth[i], max_samples=sampling[j])
            model.fit(X_train_scaled, y_train_classification)
            y_pred = model.predict(X_val_scaled)
            accuracies[i, j] = accuracy_score(y_val_classification, y_pred)
    plt.figure(figsize=(8, 4))
    sns.heatmap(accuracies, annot=True, xticklabels=sampling, yticklabels=max_depth)
    plt.title(f'Accuracy for n={n} and max depth={max_depth}')
    plt.xlabel('Sampling')
    plt.ylabel('Max depth')
    plt.show()
```
